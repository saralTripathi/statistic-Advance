   -  Random variables are classified into two main types based on the nature of their possible values:

      1.Discrete Random Variable

      2.Continuous Random Variable

3. What is the difference between discrete and continuous distributions?
   -  1. Discrete Random Variable:
        Takes on a finite or countably infinite number of distinct values.

         Typically arises in counting scenarios (e.g., number of heads in coin flips, number of customers in a store).

          Has a probability mass function (PMF) that gives the probability of each possible value.

      2. Continuous Random Variable:
         Takes on an uncountable number of values (e.g., any real number within an interval).

            Often models measurements like height, temperature, or time.

            Described by a probability density function (PDF) rather than a PMF.

            The probability of a specific single value is zero; only probability over an interval matters.

4. What are probability distribution functions (PDF)?
    -  A Probability Distribution Function describes how probabilities are distributed over the values of a random variable. The type of function used depends on whether the random variable is discrete or continuous.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
    - The Central Limit Theorem (CLT) is one of the most fundamental theorems in statistics. It states that:

      For a sufficiently large sample size, the sampling distribution of the sample mean approaches a normal distribution, regardless of the shape of the original population distribution.

      In simple terms, no matter how a population is distributed, the distribution of the sample means will be approximately normal if the sample size is large enough.

17. What are point estimates and interval estimates in statistics?
    -  In statistics, point estimates and interval estimates are used to estimate unknown population parameters based on sample data.
       
       1. Point Estimates : A point estimate is a single value used to approximate a population parameter.

       2. Interval Estimates : An interval estimate provides a range of values within which the population parameter is expected to lie, with a certain level of confidence.
18. What is the significance of confidence intervals in statistical analysis?
    -  A confidence interval (CI) is a range of values that is likely to contain the true population parameter with a certain level of confidence (e.g., 95%). It is a key tool in inferential statistics, helping us make informed conclusions about a population based on sample data.

22. What is the concept of expected value in a probability distribution?
    -  The expected value (also called mean or expectation) of a probability distribution represents the average outcome we expect over many repetitions of a random experiment. It provides a measure of the central tendency of a random variable.
